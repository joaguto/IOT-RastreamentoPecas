```python
# Importando pacote necessario
from tkinter import *
import pymysql.cursors
from datetime import datetime
import pandas as pd
import MySQLdb

# Criando a interface gráfica
class Application:
    def __init__(self, master=None):

        # Titulo
        self.fontePadrao = ("Arial", "10")
        self.primeiroContainer = Frame(master)
        self.primeiroContainer["pady"] = 15
        self.primeiroContainer.pack()

        # Valor do Qr Code
        self.segundoContainer = Frame(master)
        self.segundoContainer["padx"] = 20
        self.segundoContainer.pack()

        # Opção de escolha dos Qr Code
        self.setimoContainer = Frame(master)
        self.setimoContainer["pady"] = 5
        self.setimoContainer.pack()

        # Resposta final
        self.sextoContainer = Frame(master)
        self.sextoContainer["pady"] = 5
        self.sextoContainer.pack()

        # Resposta final
        self.descricao = Frame(master)
        self.descricao["pady"] = 5
        self.descricao.pack()

        # Resposta final
        self.lote = Frame(master)
        self.lote["pady"] = 5
        self.lote.pack()

        # Resposta final
        self.reaficoes = Frame(master)
        self.reaficoes["pady"] = 5
        self.reaficoes.pack()

        # Resposta final
        self.data = Frame(master)
        self.data["pady"] = 5
        self.data.pack()

        # Definindo o titulo
        self.titulo = Label(self.primeiroContainer, text="Interface Gráfica de Leitura Qr Code")
        self.titulo["font"] = ("Arial", "15", "bold")
        self.titulo.pack()

        # Códgio do RG
        self.qrCode = Label(self.segundoContainer, text="Código QR Code", font=self.fontePadrao)
        self.qrCode.pack(side=LEFT)

        self.qrCode = Entry(self.segundoContainer)
        self.qrCode["width"] = 30
        self.qrCode["font"] = self.fontePadrao
        self.qrCode.pack(side=LEFT)

        # Opção numero 1 - Gravar no banco de Dados
        self.autenticar = Button(self.setimoContainer)
        self.autenticar["text"] = "Gravar no Banco de Dados"
        self.autenticar["font"] = ("Calibri", "8")
        self.autenticar["width"] = 20
        self.autenticar["command"] = self.gravar_dados # Caso escolha essa opção trazer a formula
        self.autenticar.pack(side=LEFT)

        # Opção numero 2 - Trazer outras informações
        self.autenticar = Button(self.setimoContainer)
        self.autenticar["text"] = "Verificar as Informações"
        self.autenticar["font"] = ("Calibri", "8")
        self.autenticar["width"] = 20
        self.autenticar["command"] = self.informacoes # Caso escolha essa opção trazer a formula
        self.autenticar.pack(side=RIGHT)

        # Resultado final
        self.msg1 = Label(self.sextoContainer, text="", font=self.fontePadrao)
        self.msg1.pack()

        # Resultado final
        self.msg2 = Label(self.descricao, text="", font=self.fontePadrao)
        self.msg2.pack()

        # Resultado final
        self.msg3 = Label(self.lote, text="", font=self.fontePadrao)
        self.msg3.pack()

        # Resultado final
        self.msg4 = Label(self.reaficoes, text="", font=self.fontePadrao)
        self.msg4.pack()

        # Resultado final
        self.msg5 = Label(self.data, text="", font=self.fontePadrao)
        self.msg5.pack()

    # Criando função para gravar dados no banco
    def gravar_dados(self):

        connection = pymysql.connect(host='localhost',
                                     user='root',
                                     password='2305guto',
                                     db='guhring',
                                     charset='utf8mb4',
                                     cursorclass=pymysql.cursors.DictCursor)
        sql = "INSERT INTO reaficao_teste (RG, DATA_RETORNO) VALUES(%s, %s)"
        hoje = datetime.now().strftime('%d/%m/%Y, %H:%M:%S')

        # Executa query Insert
        connection.cursor().execute(sql, (str(self.qrCode.get()), hoje))
        connection.commit()
        # Encerra conexão com Mysql
        connection.close()

        self.msg1["text"] = "Dados Gravados no Banco com Sucesso"

    def informacoes(self):
        # Inputs
        mysql_cn = MySQLdb.connect(host='localhost',
                                   port=3306, user='root', passwd='2305guto',
                                   db='guhring')
        sql = "SELECT reaficao_teste.RG, SUB_LOTE, DIAMETRO_MM, DATA_RETORNO, LOTE, CLIENTE, DATA_FABRICAo, descricao, Vida_util_days, diametro_mm, comprimento_total_MM, comprimento_corte_mm, quantidade_arestas, material, TRATAMENTO_SUPERFICAL FROM guhring.reaficao_teste INNER JOIN guhring.fabricao ON reaficao_teste.RG = fabricao.RG"
        df = pd.read_sql(sql, con=mysql_cn)

        # Realizando o filtro
        df = df.loc[df.RG == self.qrCode.get()]

        # Encerrando conexão
        mysql_cn.close()

        # Coletando dados
        n_reaficoes = df.RG.count()
        descricao = df.iloc[0].descricao
        lote = df.iloc[0].LOTE
        data_fabricao = df.iloc[0].DATA_FABRICAo
        lote2 = df.iloc[0].SUB_LOTE
        diametro = df.iloc[0].DIAMETRO_MM

        self.msg1["text"] = "Lote", lote
        self.msg2["text"] = "Descricao", descricao, 'DØ', diametro
        self.msg3["text"] = 'N° de Reaficoes', n_reaficoes
        self.msg4["text"] = 'Data de Fabricação', data_fabricao
        self.msg5['text'] = 'Sublote', lote2

root = Tk()
root.title('Projeto Integrador 2021')
# Descomentar para colocar a imagem 2
#imagem2 = PhotoImage(file="empresa.png")
#z = Label(root, image=imagem2)
#z.imagem = imagem2
#z.pack()
#z.place(x=950, y=0)

Application(root)
root.mainloop()
root.update()
```
